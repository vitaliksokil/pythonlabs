```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import math
import pylab
import random
```

```python
def eps_pdf(l, x):
    return (l * math.exp(-l * x))


def normal_pdf(x, mu, sigma):
    sqrt_to_pi = math.sqrt(2 * math.pi)
    return (math.exp(-(x - mu) ** 2 / 2 / sigma ** 2) / (sqrt_to_pi * sigma))


def graph_sample_theory(list_x):
    fig, axes = plt.subplots(figsize=(8, 4), dpi=80)
    plt.hist(list_x, bins='auto', color='purple', density=True)
    plt.xlabel('Hours')
    plt.ylabel('Value')

    axes.set_xlim([0, 150])
    axes.set_ylim([0, 0.12])
    plt.plot(sorted(list_x), [normal_pdf(x, np.mean(list_x), np.std(list_x)) for x in sorted(list_x)], color='green',
             lw=4)
    plt.xlabel('Hours')
    plt.ylabel('Probability')
    plt.show()
```

```python

l = 0.0125

list_xt = range(600)
list_yt = [eps_pdf(l, x=i) for i in list_xt]

fig, axes = plt.subplots(figsize=(8, 4), dpi=80)
plt.plot(list_xt, list_yt, color='g', lw=4)
plt.xlabel('Hours')
plt.ylabel('Probability')
plt.show()
```

```python
value = 1000
n = 1

def sample_expon(value, l ,n):
    mu = 1/l
    list_x = []
    for i in range(value):
        summa = 0
        for j in range(n):
            summa += mu**2*random.expovariate(mu)
        list_x.append(summa/n)
    return list_x

list_x = sample_expon(value, l, n)

fig, axes = plt.subplots(figsize=(8, 4), dpi=80)
axes.set_xlim([0,600])
plt.hist(list_xt,bins='auto', color='purple')
plt.xlabel('Hours')
plt.ylabel('Value')
plt.show()
```

```python
fig, axes = plt.subplots(figsize=(8, 4), dpi=80)
plt.hist(list_xt,bins='auto', color='purple', density=True,label='Sample')
plt.xlabel('Mean')
plt.ylabel('Value')

axes.set_xlim([0,600])
plt.plot(list_xt, list_yt, label='Theoretical density',color='g',linestyle='-', lw=4)
plt.xlabel('Hours')
plt.ylabel('Probability')
plt.legend()
plt.show()
```

```python
w = 1
choice = (5,50,100,150,300,500)

for n in choice:
    for i in range(w):
        list_x = sample_expon(value,l,n)
        graph_sample_theory(list_x)
```
